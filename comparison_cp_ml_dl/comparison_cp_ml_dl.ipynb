```python
import pandas as pd

from sklearn.datasets import make_regression

from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
```

#CLASSIC

```python
data = make_regression(n_samples=100, noise=10, n_features=1, random_state=42)
x_data, y_data = data[0], data[1]

def classic_regressor(x): return x*45
y_pred_cp = [classic_regressor(x) for x in x_data]

plt.plot(x_data, y_pred_cp, color="r", label='classic funcion')
plt.scatter(x_data, y_data, label='actual values')
plt.legend()
plt.grid()
plt.title("Classic Programming")
plt.show()
```

#MACHINE LEARNING

```python
model_ml = LinearRegression()
model_ml.fit(x_data, y_data)
y_pred_ml = model_ml.predict(x_data)
plt.plot(x_data, y_pred_ml, color="r", label='machine learning funcion')
plt.scatter(x_data, y_data, label='actual values')
plt.grid()
plt.legend()
plt.show()
```

#DEEP LEARNING

```python
deep_regressor = MLPRegressor(max_iter=3000)
deep_regressor.fit(x_data, y_data)
y_pred_dl = deep_regressor.predict(x_data)
plt.plot(x_data, y_pred_dl, color="r", label='DEEP learning funcion')
plt.scatter(x_data, y_data, label='actual values')
plt.grid()
plt.title("DEEP prediction")
plt.legend()
plt.show()
```

```python
df = pd.DataFrame({
    "column-cp":y_pred_cp,
    "column-ml":y_pred_ml,
    "column-dl":y_pred_dl,
    "values":y_data
})
df['column-cp'] = df['column-cp'].astype("float64")
```

```python
print(mean_absolute_error(y_data, y_pred_cp))
print(mean_absolute_error(y_data, y_pred_ml))
print(mean_absolute_error(y_data, y_pred_dl))
```
